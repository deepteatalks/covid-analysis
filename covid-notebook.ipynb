### Question 1: Which state have the most number of cases?
### Question 2: Get percentage value of currentstatus from total values.

```python
import json
import pandas as pd
from pandas.io.json import json_normalize

with open('rawdata.json', 'r') as json_file:
    data = json.load(json_file)

df = json_normalize(data['raw_data'])
df.head()
```

### Converting numcases to int/float with to_numeric function

```python
df['numcases'] = pd.to_numeric(df['numcases'])
df.groupby('detectedstate')['numcases'].sum()
```

```python
df.groupby('detectedstate')['numcases'].sum().sort_values(ascending=False)
```

### Getting output of currentstatus as dict, and creating new DataFrame

```python
patient_stats_dict = df.groupby('currentstatus')['numcases'].sum().sort_values(ascending=False).to_dict()
stats_df = pd.DataFrame(patient_stats_dict.items())
stats_df.columns = ['patient_status', 'patient_count']
stats_df.head()
```

### Adding percentage column

```python
stats_df['percentage'] = (stats_df['patient_count'] * 100) /  stats_df['patient_count'].sum()
stats_df.head()
```
